```python
import random, time
    
def find_shortest_path(row):
    
    if row['node_path'] == 'NaN':
        
        global counter, segs_not_found
        start_time = time.time()
        too_long = False
        bounding_nodes = [temp_data.loc[row['index']-1,'A'],temp_data.loc[row['index']+1,'B']]
        mode = row['MODE']
        start_node = row['A']
        end_node = row['B']
        depth = 1
        max_depth = 100
        max_time = 60
        node_tree = [[start_node]]
        if row['next_nodes'] != 'NaN':
            node_tree = [node_tree[0] + row['next_nodes']]
            max_depth = 500
            max_time = 120
        while True:
            if (time.time() - start_time > 30) and too_long == False:
                print('Node pair ('+str(start_node)+', '+str(end_node)+') taking a long time to process.')
                too_long = True
            new_tree = []
            for nodes in node_tree:
                try:
                    if mode in [1,2,8,11]:
                        new_nodes = network.loc[(network['A'] == nodes[-1]) & (network['LINKC_AM'] == 42), 'B'].to_list()
                    else:
                        new_nodes = network.loc[(network['A'] == nodes[-1]) & (network['LINKC_AM'] != 42), 'B'].to_list()
                    if any(x == end_node for x in new_nodes):
                        counter += 1
                        if counter % 100 == 0:
                            print(counter,'pairs processed')
                        nodes.append(end_node)
                        return split_list(nodes)
                    if (depth >= 10) and (len(new_nodes) > 2) and (random.random() < 0.5):
                        new_nodes = []
                    for new_node in new_nodes:
                        if new_node not in nodes + bounding_nodes:
                            nodes_copy = nodes.copy()
                            nodes_copy.append(new_node)
                            new_tree.append(nodes_copy)
                except:
                    None
            node_tree = new_tree
            depth += 1
            if (depth == max_depth) or (len(node_tree) > 50000) or (time.time() - start_time > max_time):
                counter += 1
                segs_not_found.append('"A"='+str(start_node)+' OR "B"='+str(end_node))
                print('Node pair ('+str(start_node)+', '+str(end_node)+') not found at depth='+str(depth)+' and tree length='+str(len(node_tree)))
                return 'NaN'

    return row['node_path']

def split_list(l):
    new_list = []
    for i in range(len(l)-1):
        new_list.append((l[i],l[i+1]))
    return new_list
```

```python
missing_links['node_path'] = 'NaN'
missing_links['next_nodes'] = 'NaN'

counter = 1
segs_not_found = []
missing_links['node_path'] = missing_links.apply(find_shortest_path, axis=1)
if len(segs_not_found) > 0:
    print("Segments not found:")
    for s in segs_not_found:
        print(s)
```

```python
distsum = 0
total_rows = df_unique.shape[0]
rows_processed = 0
existing_nodes = {}

def find_nodes(row):
    global distsum, existing_nodes, rows_processed
    rows_processed +=1
    print('Finding nodes: %.1f' % (100 * rows_processed / total_rows) + '% processed', end='\r')
    if distsum == 0:
        existing_nodes[row['A']] = True
    distsum += row['DISTANCE']
    if row['B'] in existing_nodes:
        distsum = 0
        return row
    if row['STOPB'] == 0:
        return row
    if row['STOPB'] == 1:
        if rows_processed < total_rows:
            if df_unique.loc[row.name+1,'LINENO'] == row['LINENO']:
                if (distsum < 0.2) or \
                   ((df_unique.loc[row.name+1,'B'] in existing_nodes) and \
                   (df_unique.loc[row.name+1,'DISTANCE'] < 0.2)):
                    return row
    existing_nodes[row['B']] = True
    distsum = 0
    return row

df_unique = df_unique.apply(lambda row: pd.Series(find_nodes(row)), axis=1)

print('Finding nodes: finished           ')
```

```python
distsum = 0
links = []
stops = []
rows_processed = 0

def consolidate_links_and_nodes(row):
    global distsum, links, stops, rows_processed
    stopa = 0
    stopb = 0
    totaldist = np.nan
    alllinks = np.nan
    allstops = np.nan
    rows_processed +=1
    print('Consolidating links and nodes: %.1f' % (100 * rows_processed / total_rows) + '% processed', end='\r')
    distsum += row['DISTANCE']
    line_slice = row['geometry'].coords
    if len(links) > 0:
        line_slice = row['geometry'].coords[1:]
    links += [l for l in line_slice]
    if len(stops) == 0:
        stops.append((row['A'],Point(links[0])))
        stopa = 1
    if row['B'] in existing_nodes:
        totaldist = distsum
        alllinks = LineString(links)
        stops.append((row['B'],Point(links[-1])))
        stopb = 1
        allstops = stops
        distsum = 0
        links = []
        stops = []
    return [stopa, stopb, totaldist, alllinks, allstops]

df_unique[['STOPA','STOPB','SEGDIST','SEGLINK','SEGSTOPS']] = df_unique.apply(lambda row: pd.Series(consolidate_links_and_nodes(row)), axis=1)

print('Consolidating links and nodes: finished           ')
```

```python
def stop_nodes(group):
    global stop_list, link_list, dist_list, node_list, counter
    counter = 0
    link_list = []
    stop_list = []
    dist_list = []
    node_list = []
    group.apply(get_nodes, axis=1)
    return [list(zip(link_list, node_list, dist_list)), stop_list]
    
def get_nodes(row):
    global stop_list, link_list, dist_list, node_list, counter
    link_list.append(row['SEGLINK'])
    node_list.append(str(row['A'])+'_'+str(row['B']))
    dist_list.append(row['SEGDIST'])
    if counter == 0:
        stop_list.append(row['SEGSTOPS'][0])
    stop_list.append(row['SEGSTOPS'][1])
    counter += 1
```

```python
from shapely.geometry import Point, LineString, MultiLineString

def seg_split(row):
    point_a = train_stns.loc[train_stns['STOP_ID'] == row['STOPPATTERN'][0]['STOP_ID'], 'geometry'].values
    point_b = train_stns.loc[train_stns['STOP_ID'] == row['STOPPATTERN'][-1]['STOP_ID'], 'geometry'].values
    point_c = train_stns.loc[train_stns['STOP_ID'] == row['STOPPATTERN'][int(len(row['STOPPATTERN'])/2)]['STOP_ID'], 'geometry'].values
    train_segs['snap_dist'] = train_segs.geometry.distance(gpd.GeoSeries(list(point_a)*train_segs.shape[0]))
    train_segs['snap_dist'] += train_segs.geometry.distance(gpd.GeoSeries(list(point_b)*train_segs.shape[0]))
    train_segs['snap_dist'] += train_segs.geometry.distance(gpd.GeoSeries(list(point_c)*train_segs.shape[0]))
    closest = train_segs.loc[train_segs['snap_dist'] <= max(100, train_segs['snap_dist'].min())].reset_index(drop=True)
    closest['pos_a'] = closest.geometry.project(gpd.GeoSeries(list(point_a)*closest.shape[0]))
    closest['pos_b'] = closest.geometry.project(gpd.GeoSeries(list(point_b)*closest.shape[0]))
    try:
        closest = closest.loc[closest['pos_b'] - closest['pos_a'] > 0].iloc[0]
        backwards = False
    except:
        closest = closest.iloc[0]
        backwards = True

    line = closest['geometry']
    # First coords of line (start + end)
    coords = [l for l in line.coords]
    pointcoords = []
    # Add the coords from the points
    for i in range(len(row['STOPPATTERN'])):
        new_point = train_stns.loc[train_stns['STOP_ID'] == row['STOPPATTERN'][i]['STOP_ID'], 'geometry'].values
        pointcoords += new_point[0].coords
        coords += new_point[0].coords
    # Calculate the distance along the line for each point

    coords = list( dict.fromkeys(coords) )
    pointcoords = list( dict.fromkeys(pointcoords) )

    dists = [line.project(Point(p)) for p in coords]
    pointdists = [line.project(Point(p)) for p in pointcoords]
    # sort the coordinates
    sortpointcoords = [p for (d, p) in sorted(zip(pointdists, pointcoords))]
    coords = [p for (d, p) in sorted(zip(dists, coords))]
    if backwards:
        sortpointcoords.reverse()
        coords.reverse()

    result = []
    sublist = []

    separators = sortpointcoords[1:-1]

    for x in coords:
        if x in separators:
            if sublist:
                sublist.append(x)
                result.append(LineString(sublist))
            sublist = [x]
        else:
            sublist.append(x)
    result.append(LineString(sublist))
    return MultiLineString(result)
```

```python
offset = 200
bbox = train_stns.bounds + [-offset, -offset, offset, offset]
```

```python
hits = bbox.apply(lambda row: list(train_segs.sindex.intersection(row)), axis=1)
```

```python
import pandas as pd
import numpy as np
tmp = pd.DataFrame({
    # index of points table
    "pt_idx": np.repeat(hits.index, hits.apply(len)),
    # ordinal position of line - access via iloc later
    "line_i": np.concatenate(hits.values)
})
```

```python
# Join back to the lines on line_i; we use reset_index() to 
# give us the ordinal position of each line
tmp = tmp.join(train_segs.reset_index(drop=True), on="line_i", lsuffix="_")
# Join back to the original points to get their geometry
# rename the point geometry as "point"
tmp = tmp.join(train_stns.geometry.rename("point"), on="pt_idx", lsuffix="_")
# Convert back to a GeoDataFrame, so we can do spatial ops
tmp = gpd.GeoDataFrame(tmp, geometry="geometry", crs=train_stns.crs)
```

```python
tmp["snap_dist"] = tmp.geometry.distance(gpd.GeoSeries(tmp.point))

# Discard any lines that are greater than tolerance from points
tmp = tmp.loc[tmp.snap_dist <= 200]
# Sort on ascending snap distance, so that closest goes to top
tmp = tmp.sort_values(by=["snap_dist"])
# group by the index of the points and take the first, which is the
# closest line 
closest = tmp.groupby("pt_idx").first()
# construct a GeoDataFrame of the closest lines
closest = gpd.GeoDataFrame(closest, geometry="geometry")
```

```python
# Position of nearest point from start of the line
pos = closest.geometry.project(gpd.GeoSeries(closest.point))
# Get new point location geometry
new_pts = closest.geometry.interpolate(pos)
```
